```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

path = r'C:\Users\bwelp\Desktop\Cognixia\Cognixia Projects\Project3\Video_Games.csv'
df = pd.read_csv(path)
```

***
Part 2 TRANSFORM

```python
df
"""
Records Before any  Changes
"""
```

```python
df = df.dropna()
#Removed any records with any NaN Values in any columns
```

```python
global_sales = df[['Name','Platform','Year_of_Release','Genre','Publisher','Global_Sales','Critic_Score','User_Score','Developer','Rating']]
global_sales['Year_of_Release'] = global_sales['Year_of_Release'].astype('datetime64[D]')
global_sales['User_Score'] = global_sales['User_Score'].astype(float)
global_sales
```

***
Part 3 ANALYZE

```python
lowest_crit_scores = global_sales.query('Critic_Score == Critic_Score.min()')
lowest_crit_score = global_sales['Critic_Score'].min()

print("Lowest Critic Score:", lowest_crit_score)#shows the highest Critic Score
lowest_crit_scores #shows the list of records with the highest score
```

```python
highest_crit_scores = global_sales.query('Critic_Score == Critic_Score.max()')
highest_crit_score = global_sales['Critic_Score'].max()

print("Highest Critic Score:", highest_crit_score)#shows the highest Critic Score
highest_crit_scores #shows the list of records with the highest score
```

```python
avg_crit_score = global_sales['Critic_Score'].mean().round()
print("Average Critic Score:",avg_crit_score)
```

```python
min_user_scores = global_sales.query('User_Score == User_Score.min()')
min_user_score = global_sales['User_Score'].min()

print("Lowest User Score:", min_user_score) #shows the Lowest User Score
min_user_scores #shows the list of records with the Lowest score
```

```python
highest_user_scores = global_sales.query('User_Score == User_Score.max()')
highest_user_score = global_sales['User_Score'].max()

print("Highest User Score:", highest_user_score) #shows the Highest User Score
highest_user_scores #shows the list of records with the Highest score
```

```python
avg_user_score = global_sales['User_Score'].astype(float).mean().round()
print("Average User Score:",avg_user_score)
```

```python
avg_score_by_platform = global_sales.groupby(['Platform']).mean().sort_values(by = ['Global_Sales'], ascending= False)
print("Average Scores based on Platform(Console)")
avg_score_by_platform # 
```

```python
avg_score_by_dev = global_sales.groupby(['Developer']).mean().sort_values(by = ['Global_Sales'], ascending= False).head(20)
print("Average Scores based on Developer")#Top 20 Developers with the highest global sales with critic score and user scores.
avg_score_by_dev
```

The top 20 developers taht sold the most have a higher chance of getting a higher Critic Score and User Score as more data is provided

```python
game_count = global_sales['Platform'].value_counts().tolist()
label = list(global_sales['Platform'].value_counts().index.values)
fig, ax = plt.subplots()
fig.set_size_inches(10,10)
ax.pie(game_count,labels = label,autopct='%1.1f%%')
ax.set_title('Games Produced Based on Platform')

plt.show()
```

The Graph above represents the games made from the specific platform ie. the Playstation 2 taking a 16.7 percentage chunk from the game pool.

```python
usr_score = global_sales['User_Score'].tolist()
cscore = global_sales['Critic_Score'].tolist()
u_score = []
for i in range(len(usr_score)):
    u_score.append(usr_score[i] * 10) 

fig, ax = plt.subplots()
fig.set_size_inches(15,15)
plt.scatter(u_score,cscore)

line = mlines.Line2D([0, 1], [0, 1], color='red')
transform = ax.transAxes
line.set_transform(transform)
ax.add_line(line)

plt.xlabel('User Score')
plt.ylabel('Critic Score')
plt.title('Ciritc Score vs User Score Scatterplot')

plt.show()



```

Graph above shows the comparison of User Scores and Critic Score. The line represents the User score and critic score are the same value thus the game score is accurate from the critics and from players that played the game.
